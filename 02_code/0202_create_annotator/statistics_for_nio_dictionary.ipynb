```python
import xmltodict
import collections
import altair as alt
from vega_datasets import data
import pandas as pd
```

```python
# read .xml as a dictionary 
## input format:
# <?xml version="1.0" ?>
# <synonym>
#     <token id="NDDO:NDDO_00000021" canonical="visit">
#         <variant base="visit"/>
#     </token>
#     <token id="NDDO:NDDO_00000022" canonical="diagnostic data">
#         <variant base="diagnostic data"/>
#     </token>
# </synonym>

## output format:
#  [OrderedDict([('@id', 'NDDO:NDDO_00000021'),
#               ('@canonical', 'visit'),
#               ('variant', OrderedDict([('@base', 'visit')]))]),
#  OrderedDict([('@id', "NDDO:NDDO_10002233"),
#               ('@canonical', "Boston Naming Test"),
#               ('variant', [OrderedDict([('@base', 'Boston Naming Test Scale')]), 
#                            OrderedDict([('@base', 'Boston Naming Test')])])])]

xml_path = "/Users/yidesdo21/Projects/inputs/dictionary/nio.xml"

with open(xml_path) as f:
    nio_xml = f.read()

token_cano = dict()  # {"token_id": "canonical"}
nio_parsed = xmltodict.parse(nio_xml)
```

```python
nio_dict = nio_parsed["synonym"]["token"]
print("number of token ids:", len(nio_dict)) # number of token ids 
```

```python
# nio_dict
```

```python
## for each token id, how many variants does it have
id_var = dict()
for pair in nio_dict:
    variants = pair["variant"]
    token_id = pair["@id"]
    if type(variants) is list:
        id_var[token_id] = len(variants)
#         print(pair)
#         print(variants)
#         print("-------")
    else:
        id_var[token_id] = 1
```

```python
for token_id, cnt in id_var.items():
    if cnt > 5:
        print(token_id,cnt)
```

```python
# turn the dictionary into a pandas dataframe
id_var_pd = pd.Series(id_var, name='num_of_variants')
id_var_pd.index.name = 'token_id'
id_var_df = id_var_pd.to_frame().reset_index()
```

```python
alt.Chart(id_var_df).mark_bar().encode(
    alt.X("num_of_variants:Q", bin=False),
    y='count()',
    tooltip=['num_of_variants', "count()"]
).interactive()
```

```python
## for the variants
## for each variant, how many token ids does it correspond to? 
#  [OrderedDict([('@id', 'NDDO:NDDO_00000021'),
#               ('@canonical', 'visit'),
#               ('variant', OrderedDict([('@base', 'visit')]))]),  <--- This is what is examined here
#  OrderedDict([('@id', "NDDO:NDDO_10002233"),
#               ('@canonical', "Boston Naming Test"),
#               ('variant', [OrderedDict([('@base', 'Boston Naming Test Scale')]), <--- This is what is examined here
#                            OrderedDict([('@base', 'Boston Naming Test')])])])]
# the number of distinct variants 
all_variants = list()
distinct_variants = set()
for pair in nio_dict:
    variants = pair["variant"]
    if type(variants) is list:
        for variant in variants:
#             print(variant["@base"])
            var = variant["@base"]
            all_variants.append(var)
            distinct_variants.add(var)
    else:
        var = variants["@base"]
        all_variants.append(var)
        distinct_variants.add(var)
#         print(variants["@base"])
#     print("-------")
    
```

```python
print("number of all variants:", len(all_variants))
```

```python
print("number of distinct variants:", len(distinct_variants))
```

```python
cnt, dup_cnt = 0,0
all_var, dup_var = list(), list()
dup, single, others = list(), list(), list()
for var, count in collections.Counter(all_variants).items():
    cnt += count
    all_var.append(var)
#     print(var,count)
    if count > 1:
        dup_cnt += count
        dup_var.append(var)
        dup.append((var,count))
    elif count == 1:
        single.append((var,count))
    else:
        others.append((var,count))
#         print(var, count)        
```
