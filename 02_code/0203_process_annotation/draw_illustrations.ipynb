```python
import networkx as nx
import nx_altair as nxa   # altair can't add the label for the nodes
import matplotlib.pyplot as plt
```

```python
### when nodes are in both the testing set and the training set
# generate a graph
G = nx.Graph()
edges = list()
edge_g1 = [('A','B'), ('B','D'), ('A','C'), ('C','D')]
edge_g2 = [('E','F'), ('F','G'), ('E','G')]
edges.extend(edge_g1)  
edges.extend(edge_g2)  
G.add_edges_from(edges)
```

```python
## graph from the pre-cutoff dataset
nx.draw(G, with_labels=True, font_weight='bold', font_color="white")
```

```python
G_post = nx.Graph(G)
edge_g3 = [('A', 'F')]  # new edges in the post-cutoff dataset
G_post.add_edges_from(edge_g3)
```

```python
## graph from the post-cutoff dataset
nx.draw(G_post, with_labels=True, font_weight='bold', font_color="white")
```

```python
## when new nodes are in the testing set but not in the training set
## This can never work: we can never predict the missing link between A and F with the graph from the training dataset
## because the new nodes are never in the training graph
# the graph from the training dataset
G_new = nx.Graph()
G_new.add_edges_from(edge_g1)
nx.draw(G_new, with_labels=True, font_weight='bold', font_color="white")
```

```python
# thee graph from the testing dataset
nx.draw(G, with_labels=True, font_weight='bold', font_color="white")
```
